## Drop outliners and plot

```python
data = data[(data.start_lat >= 41.6) & (data.start_lat <= 42.1) & 
            (data.start_lng >= -87.9) & (data.start_lng <= -87.5)]
data.head(2)
```

```python
fig, ax = plt.subplots(figsize=(12, 6), ncols=2)
ax[0].scatter(data.start_lng, data.start_lat, c='red', alpha=.4)
ax[1].scatter(data.end_lng, data.end_lat, c='blue', alpha=.4)
plt.tight_layout()
plt.show()
```
