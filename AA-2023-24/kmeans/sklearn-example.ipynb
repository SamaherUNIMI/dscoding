## Example: transformers and classifiers

```python
import numpy as np
```

```python
data = load_iris()
X = data.data
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
Xt = scaler.transform(X)
```

```python
scaler.transform(np.array([[1, 2, 3, 4], [4, 5, 6, 7]]))
```

```python
pca = PCA(n_components=2)
pca.fit(X)
Xp = pca.transform(X)
```

```python
pipeline = [PCA(n_components=2), MinMaxScaler()]
T = X
for algorithm in pipeline:
    T = algorithm.fit_transform(T)
```
